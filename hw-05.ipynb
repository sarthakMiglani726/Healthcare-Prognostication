##Visualization
Average Risk of Dementia by Age Group

```python
# Extracting the relevant columns for age groups and dementia risk
age_group_columns = [col for col in data1.columns if col.startswith('age group:')]
age_group_data = data1[age_group_columns]
dementia_risk = data1['predicted risk of Dementia']

# Preparing the data for the plot
# Summing up the number of people in each age group
age_group_totals = age_group_data.sum()

# Calculating the average risk of dementia for each age group
average_risk_per_age_group = {}
for col in age_group_columns:
    # Getting the indices where the age group is 1 (i.e., the patient belongs to that age group)
    indices = age_group_data[age_group_data[col] == 1].index
    # Calculating the average risk for this age group
    average_risk_per_age_group[col] = dementia_risk[indices].mean()

# Converting to a format suitable for plotting
age_groups = list(average_risk_per_age_group.keys())
average_risks = list(average_risk_per_age_group.values())

# Creating a horizontal bar plot for average risk of dementia by age group

plt.figure(figsize=(12, 8))

# Plotting horizontal bars
plt.barh(age_groups, average_risks, color='darkorange')

# Adding labels to each bar for clarity
for index, value in enumerate(average_risks):
    plt.text(value, index, f'{value:.4f}', va='center', ha='left', fontsize=8)

# Enhancing the overall graph aesthetics
plt.ylabel('Age Group', fontsize=14, labelpad=15)
plt.xlabel('Average Risk of Dementia', fontsize=14, labelpad=15)
plt.title('Average Risk of Dementia by Age Group', fontsize=16, pad=20)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

The horizontal bar plot visualizes the average risk of dementia for various age groups. Each bar length represents the magnitude of risk, with annotated values offering precise risk percentages. The age group "90 - 94" has the highest average risk at approximately 0.2861, indicating a strong correlation between advancing age and the risk of developing dementia. In contrast, younger age groups exhibit significantly lower risks, with values close to 0. The data underscore the increased average risk associated with older age groups, emphasizing the need for age-specific approaches to dementia risk assessment and intervention.

###Coding Categorical Variables
In this dataset, columns such as age_group and sex that were originally categorical have been converted into numerical form. For example, different age groups and sexes are now represented by unique numerical values.

```python
# Creating a copy of the data_cleaned dataset for transformations
data_encoded = data.copy()

# Initializing a Label Encoder
le = LabelEncoder()

# Identifying categorical columns for label encoding
categorical_cols = data_encoded.select_dtypes(include=['object']).columns

# Applying Label Encoding to each categorical column
for col in categorical_cols:
    data_encoded[col] = le.fit_transform(data_encoded[col])

# Display the first few rows of the transformed dataset
data_encoded
```

###Data Transformation

* Feature Scaling:
The numerical features of the dataset were standardized using the StandardScaler from scikit-learn. This process rescales each feature to have a mean of 0 and a standard deviation of 1, which is crucial for algorithms that are sensitive to the scale of the data, like gradient descent.

* Variable Transformation:
A log transformation (using log1p, which applies log(1+x) to avoid issues with log(0)) was applied to the numerical data to address potential skewness in the distribution of variables. This transformation is particularly useful for reducing skewness in variables, which can lead to better performance in various models.

After the log transformation, a check was conducted to ensure that there were no infinite values resulting from this transformation. No infinite values were detected.

```python
# Applying log transformation to reduce skewness in the data
# We add 1 to avoid issues with log(0)
log_transformed_data = np.log1p(data_encoded)

# Checking for any infinity values after log transformation
inf_check = np.isinf(log_transformed_data).values.sum()

# Displaying the log-transformed data
log_transformed_data
```

##Ordinary Least Squares (OLS) Regression

###Assumption Checks
* Linearity Check (Observed vs. Predicted): The scatter plot does not show a perfect linear relationship, indicating potential non-linearity or the need for different/additional variables to capture a linear relationship.

* Independence Check (Durbin-Watson Test): The Durbin-Watson statistic is 2.16, which is close to 2, suggesting that the residuals are independent.

* Homoscedasticity Check (Residuals vs. Predicted values): The scatter plot of residuals versus predicted values displays some pattern, indicating that homoscedasticity may not hold.

* Normality of Residuals (Q-Q Plot): The Q-Q plot shows significant deviation from the line, particularly at the tails, suggesting that the residuals are not normally distributed. This is confirmed by the Shapiro-Wilk test, which returned a p-value of 0.0, indicating non-normality of residuals.

```python
# Selecting a few key independent variables for simplicity
# Here, we arbitrarily choose a few columns. For a comprehensive analysis, more rigorous feature selection should be performed.
independent_vars = ['ageCount', 'sexCount', 'dyspnea_in_prior_year', 'rheumatoid_arthritis_in_prior_year', 'seizure_in_prior_year', 'sepsis_in_prior_year']
X = log_transformed_data[independent_vars]
y = log_transformed_data['predicted_risk_of_dementia']

# Adding a constant to the model (the "intercept")
X = sm.add_constant(X)

# Fitting the OLS model
model = sm.OLS(y, X).fit()

# Residuals
residuals = model.resid

# Checking assumptions
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Linearity check: Scatter plot of observed vs. predicted values
predicted = model.predict(X)
sns.scatterplot(x=predicted, y=y, ax=axes[0, 0])
axes[0, 0].set_title("Linearity Check: Observed vs. Predicted")
axes[0, 0].set_xlabel("Predicted Values")
axes[0, 0].set_ylabel("Observed Values")

# Independence check: Durbin-Watson Test
# Values of the test statistic close to 2 suggest independence
durbin_watson_stat = sm.stats.durbin_watson(residuals)
axes[0, 1].text(0.1, 0.5, f'Durbin-Watson Statistic: {durbin_watson_stat}', fontsize=12)
axes[0, 1].axis('off')

# Homoscedasticity check: Residuals vs. Predicted values
sns.scatterplot(x=predicted, y=residuals, ax=axes[1, 0])
axes[1, 0].set_title("Homoscedasticity Check: Residuals vs. Predicted")
axes[1, 0].set_xlabel("Predicted Values")
axes[1, 0].set_ylabel("Residuals")
axes[1, 0].axhline(y=0, color='r', linestyle='--')

# Normality of residuals: Q-Q plot
sm.qqplot(residuals, line='s', ax=axes[1, 1])
axes[1, 1].set_title("Normality of Residuals: Q-Q Plot")

plt.tight_layout()
plt.show()

# Shapiro-Wilk normality test on residuals
shapiro_test = stats.shapiro(residuals)
shapiro_test_stat, shapiro_p_value = shapiro_test

shapiro_test_stat, shapiro_p_value, durbin_watson_stat


```

The dataset does not fully meet the OLS assumptions, especially in terms of linearity, homoscedasticity, and normality of residuals.
